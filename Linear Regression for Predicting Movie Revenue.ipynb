2. Gather Data

```python
#Select data sort acc to rank
#format in excel sheet => cntrl + 1
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
data=pd.read_csv('cost_revenue_clean.csv')
```

```python
data
```

```python
data.describe()
```

```python
X = DataFrame(data,columns=['production_budget_usd'])  #Budgets
Y = DataFrame(data,columns=['worldwide_gross_usd'])    #Revenue
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title('Film Budget Vs Global Revenue')
plt.xlabel('Production Budget')
plt.ylabel('Film Revenue')
plt.ylim(0,3000000000)
plt.show()
```

![title](Scr.png)

![title](Scr2.png)

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X,Y)
```

```python
regression.coef_ #Slope Coefficient
```

```python
regression.intercept_ #intercept
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)

plt.plot(X,regression.predict(X),color='red',linewidth=3)

plt.title('Film Budget Vs Global Revenue')
plt.xlabel('Production Budget',color='white')
plt.ylabel('Film Revenue',color='white')
plt.ylim(0,3000000000)
plt.show()
```

R2 goodness of fit

```python
    regression.score(X,Y) #statistical measure of how close the data are to the fitted regression line
```
